Second major type of supervised learning algorithm is ***Classification***

example: Breast cancer detection

the machine learning system try to find out if the tumor is maligmant or benign, the dataset may have tumors with different sizes.

here x = size of tumor, y indicate the danger of the tumor.

there is 2 possible output 1 and 0, this what make it a classification, we can plot the output on a line.
