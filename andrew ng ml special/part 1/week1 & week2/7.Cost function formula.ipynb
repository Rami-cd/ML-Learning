<h2>Cost function</h2>

m is the number of training examples, with more training examples m is larger, and the cost function will calculate a bigger number, so to build a cost function that doesn't automatically get bigger as the training set gets larger, by convention we conputes the average square error instead of the total square error by dividing by 2m (we can without the 2, it's here to make the calculation easyer)
