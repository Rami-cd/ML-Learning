```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
```

```python
x = np.arange(0,20,1)
y = np.cos(x/2)
x_2 = x**2
x_3 = x**3
x_4 = x**4
data_set = pd.DataFrame({'x': x,
                         'x_2': x_2,
                         'x_3': x_3,
                         'x_4': x_4})
data_set.head()
```

```python
data_set.iloc[0, 0]
```

```python
y_p = np.zeros(20)
w = [0.3241, 0.7003, 0.5211, 0.0421]
for i in range(len(data_set)):
    y_p[i] = np.dot(data_set.iloc[i], w)
```

```python
plt.plot(x, y_p)
```
