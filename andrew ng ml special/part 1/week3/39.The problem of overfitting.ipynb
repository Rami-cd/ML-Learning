what is overfitting ?

<img src="images/101.png">

we have this dataset, if we try to fit a straight line but it doesn't fit the data well so thid model is ***underfitting***, another term for it is ***high bias***.

another way to think of it that the learning algorithm have a very strong preconception (bias) that the housing prices going to be a linear function.

in the second one the model is better and have a good prediction.

if we have a new data on the algorithm this model will have a good prediction, this ic called generalization which mean that the model make good prediction even on new data.

in this third example we use multiple polynomial.

it's very good when dealing with the training data, the cost is low but it's not good at predicting data, the technical term is ***overfitting*** or ***high variance*** (even the smallest change can affect the prediction).

so the one in the middle is the right one.

<h2>adressing overfitting</h2>

<img src="images/103.png">

one solution is to get more data if more data exist.

another solution is to not use so many polynomial features.

<img src="images/104.png">

in this example we have so many features, we can use a subset of the most relevant features ***feature selection***, the disadvantage is that the algorithm is getting rid of some features, which are actually useful.

<h2>Regularization</h2>

<img src="images/105.png">

here we can elminate a feature (set parameter to 0).

but with ***Regularization***, we can reduce the inpact of the features without eleminating them.

so it let you keep the features without them having a big effect on the data (regularize w not b).

So these are the ways to deal with overfitting.
